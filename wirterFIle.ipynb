```python
import serial
import csv

porta_serial = 'COM3' 
baud_rate = 115200

ser = serial.Serial(porta_serial, baud_rate)

# Cria e abre o arquivo CSV
with open('dataset_motor.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Tempo_ms", "RPM", "Corrente_A"]) # Cabeçalho

    print("Coletando dados... Pressione Ctrl+C para parar.")
    try:
        while True:
            linha = ser.readline().decode('utf-8').strip()
            if linha:
                dados = linha.split(',')
                writer.writerow(dados)
    except KeyboardInterrupt:
        print("\nColeta finalizada! Arquivo 'dataset_motor.csv' gerado.")
        ser.close()
```
